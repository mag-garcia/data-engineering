*Agregamos el nombre de las plataformas de cada dataframe, un ID para identificarlo*

```python
dataframes = [amazon_df,disney_df,hulu_df,netflix_df]
platforms = ['Amazon', 'Disney', 'Hulu', 'Netflix']

for n, dataframe in enumerate(dataframes):
    dataframe['platform'] = platforms[n]

    
```

*Unimos los dataframes en un solo dataframe y quitamos las columnas que no vamos a utilizar*

```python
#df = amazon_df.append([disney_df,hulu_df,netflix_df],ignore_index=True)
#The frame.append method is deprecated and will be removed from pandas in a future version. Use pandas.concat instead.

df = pd.DataFrame()
df = pd.concat([amazon_df,disney_df,hulu_df,netflix_df],ignore_index=True)
df.drop (columns=['country','description','date_added', 'director','rating','show_id'], inplace= True)
df
```

*Eliminamos los duplicados y contamos los nulos*

```python
df.drop_duplicates(inplace=True)
```

```python
df.isnull().sum()  #df.isnull().mean()
```

*Separamos los datos de la columna 'duration' en 'duration' y 'duration_type' (unidad de medida), ya que las duraciones de películas se presentan en minutos y las series en temporadas*

```python
#creamos un nuevo df con la columna y la separamos en dos

split_duration = df.duration.str.split(" ", n=1, expand=True)
split_duration.rename(columns={0:'duration', 1:'duration_type'}, inplace= True)
```

```python
#llenamos los nulos en ambas columnas segun corresponda

split_duration['duration'].fillna(0, inplace=True)
split_duration['duration_type'].fillna('No data', inplace=True)
```

```python
split_duration.isnull().sum()
```

```python
#integramos las nuevas columnas al df

df['duration_2'] = split_duration['duration']
df['duration_type'] = split_duration['duration_type']
df.drop(columns=['duration'], inplace=True)
df.rename(columns={'duration_2':'duration'}, inplace=True)


df['duration'] = df['duration'].astype('int64')  #transformamos los datos de duración a integer
df
```

*Llenamos los datos faltantes del resto del dataframe*

```python
df.fillna('No data', inplace=True)
```

```python
df
```

```python
df.isnull().sum()
```

```python
# Se crea una columna con un indice nuevo ya que la columna show_id que dropeamos al principio tenía datos repetidos
df["id"] = df.index+1
df
```

```python
#normalizamos los datos pasando a minúsculas para facilitar las consultas

df['type'] = df['type'].str.lower()
df['listed_in'] = df['listed_in'].str.lower()
df['platform'] = df['platform'].str.lower()
df['duration_type'] = df['duration_type'].str.lower()
```

```python
#reordenamos las columnas reasignando el dataframe
df = df[['id', 'title', 'platform', 'type', 'listed_in','release_year', 'duration', 'duration_type', 'cast']]

#tambien se puede usar reindex
```

*Creación y prueba de funciones*

```python
def actores(plataforma, anio):
    cast = df[['platform','release_year','cast']]
    actores_list = cast.groupby(['platform','release_year'])['cast'].apply(lambda x: ','.join(x))

    actores = actores_list[plataforma][anio]
    actores = actores.split(',')
    actores_set = set(actores)

    masrepetido = ('', 0) 

    for actor in actores_set:
        cant = actores.count(actor)

        if actor != 'No data':
            if cant > masrepetido[1]:
                masrepetido = (actor,cant)

        
    return masrepetido
        
actores('netflix',2018)

```

```python
def plataforma (plat):
    movies = df.loc[(df['platform']==plat) & (df['type']=='movie')].shape
    series = df.loc[(df['platform']==plat) & (df['type']=='tv show')].shape
    
    movies_cant = movies[0]
    series_cant = series[0]
       

    print(f'En la plataforma {plat} hay {movies_cant} peliculas y {series_cant} series')

plataforma('amazon')
```

```python
def max_duracion (plat, year, dur):
  plat_year_dur = df.loc[(df['platform']==plat) & (df['release_year']==year) & (df['duration_type']==dur)] 
  
  plat_year_dur['duration'] = plat_year_dur['duration'].astype('float64')
  
  maximo = plat_year_dur['duration'].max()

  title = plat_year_dur['title'].loc[plat_year_dur['duration'] == maximo].values[0]

      
  if dur == 'min':
    return {f'La mayor duracion en la plataforma {plat} la tiene la pelicula {title} con {maximo} minutos'}
  else: 
    return {f'La mayor duracion en la plataforma {plat} la tiene la serie {title} con {maximo} temporadas.'}

max_duracion('netflix', 2019, 'min')
```

```python
def get_listedin (genre):
    genre_df = df.loc[(df['listed_in'].str.contains(genre))&(df['listed_in'].str.contains(genre+' ')==False)]
    plat_genre = genre_df['platform'].value_counts()
    plat_genre = plat_genre[plat_genre == plat_genre.max()].index[0]
    rep_genre = genre_df['platform'].value_counts().max()
    message = f'La plataforma con más series o películas del género {genre} es {plat_genre} con {rep_genre} repeticiones'

    return {message}

get_listedin('horror')
```
